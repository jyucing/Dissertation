```python
import json
from urllib.request import urlopen, quote
import requests,csv
import pandas as pd
```

# Residential Compound

Data can be downloaded from https://drive.google.com/drive/folders/1GOhMd0UbGnod2Kgtyw5k8_3KYFlEBHT0?usp=sharing

```python
xiaoqu = pd.read_csv("Data/xiaoqu/XiaoquPrice.csv", encoding="utf-8")
```

```python
xiaoqu
```

```python
address = xiaoqu['地址'].tolist()
address
```

```python
def getlnglat(address):
    url = 'http://api.map.baidu.com/geocoding/v3/'
    output = 'json'
    ak = 'c9Tx7ZnF6dO2n6UmiwpVDOqTgaIJIBO2' 
    add = quote(address) 
    uri = url + '?' + 'address=' + add + '&output=' + output + '&ak=' + ak
    req = urlopen(uri)
    res = req.read().decode() # Decodes strings of other encodings into unicode
    temp = json.loads(res) # Parse json data
    return temp
```

```python
# Open the output CSV file
with open('Data/Xiaoqu/XiaoquPOI3.csv', 'w', encoding='utf-8', newline='') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(["address", "lng", "lat"])
    
    for i in address:
        try:
            response = getlnglat(i)
            if 'result' in response and 'location' in response['result']:
                lng = response['result']['location']['lng']
                lat = response['result']['location']['lat']
                str_temp = [i, lng, lat]
                csv_writer.writerow(str_temp)
            else:
                # No location information found
                str_temp = [i, 'N/A', 'N/A']
                csv_writer.writerow(str_temp)
        except Exception as e:
            # Record the address where the error occurred
            str_temp = [i, 'Error', 'Error']
            csv_writer.writerow(str_temp)
```
